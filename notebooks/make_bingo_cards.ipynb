# Example BINGO Sheets

Load some example phrases to use when creating BINGO sheets.

```python
# open and process phrase sheet
fname = module_path+'bingo_items_example.txt'
with open(fname) as f:
    phrases = f.readlines()
phrases = [x.strip() for x in phrases] 
phrases = [x.replace('(new line)','\n') for x in phrases] 
print("number of phrases:",len(phrases))
```

```python
# create a random bingo sheet
card = BingoCard(phrases=phrases)
```

```python
# create and save random bingo sheets
Ncards = 10
for i in range(0, Ncards):
    card = BingoCard(phrases=phrases)
    card.fig.savefig(module_path+'notebooks/bingo_sheets/'+str(i)+'_sheet.pdf')
    plt.close(card.fig)
```
